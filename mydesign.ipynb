# PLanet Design Explorer
Compare a within-subjects and between-subjects design using `planet_ui`.
Each `show()` / `compare()` call opens a GUI window and **blocks the cell** until you close it.

```python
from planet import *
from planet_ui import show, compare
```

```python
treatment = ExperimentVariable(name="treatment", options=["control", "experimental"])
```

```python
design1 = (
    Design()
    .within_subjects(treatment)
    .counterbalance(treatment)
)

design2 = (
    Design()
    .between_subjects(treatment)
    .counterbalance(treatment)
)
```

```python
# Show a single design — analysis + plans + assignment
show(design1, units=2, label="Within-Subjects")
```

```python
# Compare two designs side-by-side
compare(design1, design2, label1="Within-Subjects", label2="Between-Subjects", units=2)
```
